```python
class MRIDataset(data.Dataset):
    
    def __init__(self, root_dir: str, labels: dict, transform):
        
        self.root_dir = root_dir
        self.labels = labels
        self.filenames = sorted(os.listdir(self.root_dir))
        self.transform = transform
    
    def __getitem__(self, idx: int):
        
        if isinstance(idx, int):
            filename = self.filenames[idx]
        img_name = os.path.join(self.root_dir, filename)
        image = nib.load(Path(img_name)).get_fdata()

        image = self.__normalize_intensity(image)
        
        image_label = self.labels.get(filename)

        if not image_label:
            image_label = np.nan

        if self.transform:
            image = self.transform(image)
        
        return torch.tensor(image), torch.tensor(image_label)
    
    def __len__(self):
        return len(self.filenames)

    def __normalize_intensity(self, image):

        pixels = image[image > 0]
        mean = pixels.mean()
        std = pixels.std()

        result = (image - mean) / std
        
        ### Preventing values that had the value == mean before the normalization
        ### From vanishing
        result_random = np.random.normal(0, 1, size=image.shape)
        result[image == 0] = result_random[image == 0]

        return result
```
